```python
import torch

from src.eval import eval_model, display_model_results
from src.train import train_model
from src.models import FCNNBreastCancer, CNNBreastCancer, ResNetBreastCancer

%load_ext autoreload
%autoreload 2
```

```python
import yaml


def read_yaml(file_path):
    with open(file_path, "r") as file:
        return yaml.safe_load(file)


config = read_yaml("config.yaml")
```

```python
fcnn_config = config["model_fcnn"]

model_dir = train_model(
    FCNNBreastCancer,
    fcnn_config["lr"],
    fcnn_config["batch_size"],
    fcnn_config["epochs"],
    config["model_path"],
    config["data_path"],
    config["device"],
)

eval_model(FCNNBreastCancer, model_dir)
display_model_results(model_dir)
```

```python
cnn_config = config["model_cnn"]

model_dir = train_model(
    CNNBreastCancer,
    cnn_config["lr"],
    cnn_config["batch_size"],
    cnn_config["epochs"],
    config["model_path"],
    config["data_path"],
    config["device"],
)
eval_model(CNNBreastCancer, model_dir)

display_model_results(model_dir)
```

```python
rcnn_config = config["model_resnet"]

model_dir = train_model(
    ResNetBreastCancer,
    rcnn_config["lr"],
    rcnn_config["batch_size"],
    rcnn_config["epochs"],
    config["model_path"],
    config["data_path"],
    config["device"],
)
eval_model(ResNetBreastCancer, model_dir)

display_model_results(model_dir)    
```
